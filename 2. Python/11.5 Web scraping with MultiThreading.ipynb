## Real world use case of Multi-threading

Web scraping is one of the real world applications of multithreading where we will be sending multiple network requests which are considered as I/O bound because they spend a lot of time waiting for responses from servers. Multithreading can significantly improve the performance by allowing web pages to be fetched concurrently.

```python
import threading
import requests
from bs4 import BeautifulSoup
import time
```

```python
# URLs for which we are planning to perform web scraping

urls = [
    'https://docs.langchain.com/oss/python/deepagents/overview',
    'https://docs.langchain.com/oss/python/deepagents/human-in-the-loop',
    'https://docs.langchain.com/oss/python/deepagents/long-term-memory'
]
```

```python
def fetch_content(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content,'html.parser')
    print(f'Fetched {len(soup.text)} characters from {url}')
```

```python
# Time taken for sequential execution

t = time.time()

for url in urls:
    fetch_content(url)

finished_time = time.time() - t
print(f"Sequential web scraping took : {finished_time}")
```

```python
# Time taken with multithreading 

t = time.time()

threads = []

for url in urls:
    thread = threading.Thread(target=fetch_content,args=(url,))
    threads.append(thread)
    thread.start()

for thread in threads:
    thread.join()

finished_time = time.time() - t

print(f"All web pages fetched using multithreading in {finished_time}")
```
