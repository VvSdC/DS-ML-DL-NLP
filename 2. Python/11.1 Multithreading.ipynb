## Python multithreading and multiprocessing

- Program: A program is a sequence of instructions written in a programming language.

- Process: A process is simply an instance of a program that is being executed. Processes share separate memory space so that once process cannot corrupt another process.

- Thread: A thread is a unit of execution within a process. Every thread in a process will have it's own stack and register.

# Implementing multi-threading

When to use multi-threading?
- I/O bound tasks: Tasks that spend more time waiting for I/O operations
- Concurrent execution: When we want to improve the throughput of application by performing multiple operations concurrently.

```python
import threading
import time
```

```python
def print_number():
    for i in range(5):
        time.sleep(2)
        print(f"Number : {i}")

def print_letter():
    for letter in "abcde":
        time.sleep(2)
        print(f"Letter : {letter}")
```

```python
t = time.time()
print_number()
print_letter()
finished = time.time() - t

print(f"Time taken to finish using single thread: {finished}")

# As of now we have a single thread program that waits for print_number() to complete first before even starting the print_letter() function
```

```python
# Using multi-threading

# Creating threads
t1 = threading.Thread(target=print_number)
t2 = threading.Thread(target=print_letter)

t = time.time()
# Starting execution of threads
t1.start()
t2.start()

# Waiting for threads to complete
t1.join()
t2.join()

finished_time = time.time() - t
print(f"Time taken using multi-threading : {finished_time}")
```
