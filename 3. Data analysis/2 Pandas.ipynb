## Pandas DataFrame And Series

Pandas is a powerful data manipulation library in Python, widely used for data analysis and data cleaning. It provides two primary data structures:
- Series (One-dimensional array-like object)
- DataFrame (Two-dimensional, size-mutable and potentially heterogeneous tabular data structure with labeled indexes)

```python
import pandas as pd
```

### Series
A pandas series is a one-dimensional array-like object that can hold any data type. It is similar to a column in the table.

```python
# Converting a list into series
data = [1,2,3,4,5,6]
series = pd.Series(data)

print(f"Pandas series is : \n{series}")
print(f"\nData type of pandas series is : {type(series)}")
```

```python
# Converting a dictionary into series
data = {'a':1 , 'b':2 , 'c':3 , 'd': 4}
series = pd.Series(data)

print(f"Series obtained from dictionary is : \n{series}")
```

```python
# Combining data and index lists
data= [10,20,30,40]
index = ['a','b','c','d']

series = pd.Series(data,index = index)
print(f"List {data} indexed with {index} is : \n{series}")
```

### DataFrame

```python
# Creating a dataframe from a dictionary of list

data = {
    'Name': ['Person1','Person2','Person3'],
    'Age': [20,30,40],
    'Location' : ['HYD','BLR','MAS']
}
dataframe = pd.DataFrame(data)
print(f"DataFrame obtained from dictionary of list values is : \n{dataframe}")
print(f"\nData type of DataFrame is : {type(dataframe)}")
```

```python
# Creating a dataframe from a list of dictionaries
data = [
    {
        'name':'Person1',
        'age':20,
        'location':'HYD'
    },
    {
        'name':'Person2',
        'age':30,
        'location':'BLR'
    },
    {
        'name':'Person3',
        'age':40,
        'location':'MAS'
    }
]

dataframe = pd.DataFrame(data)
print(f"Dataframe obtained from a list of dictionaries is : \n{dataframe}")
```

```python
# Reading csv file

df = pd.read_csv('employees.csv')
print(f"First 5 rows of data is : \n{df.head(5)}")
```

### Accessing data from dataframe

```python
# The dataframe is a 2-dimensional data structure
# When we get a single column from a dataset, it will be a series

print(f"The values in name column of dataframe is :\n{df['name']}")
print(f"The data type of name column in dataframe is : {type(df['name'])}")
```

```python
# Using loc
df.loc[0]
```

```python
# Using iloc
df.iloc[0]
```

```python
# Accessing a specified element using at
# Accessing name value from index 1 of dataframe

df.at[1,'name']
```

```python
# Accessing a specified element like we do for multi-dimensional matrix using iat
# Accessing element at second row and second column of dataframe
df.iat[1,1]
```

### Data Manipulation

```python
# Current dataframe

df
```

```python
# Adding location column data

df['Location'] = ['HYD','BLR','DEL','MUM','TPTY','VSP','CCT','MAS']
df
```

```python
# Dropping a column in dataframe
# Axis = 1 checks for that particular column name while Axis = 0 will check for row index
# A normal drop is a temporary operation and will not effect original dataframe

print(f"When we drop the age column : \n{df.drop('age',axis = 1)}")
print(f"\nThe original dataframe : \n{df}")
```

```python
# To make drop permanent and persist in dataframe, we must use inplace = True

df.drop('age',axis=1,inplace=True)
print(f"\nOriginal dataframe is : \n{df}")
```

```python
# Incrementing everyone's salary by 15000

df['salary'] = df['salary'] + 15000
print(f"Dataframe after incrementing salary is : \n{df}")
```

```python
# Dropping a row
# As mentioned in the top, by default the axis parameter value will be 0 so the drop function will look at rows
# And to make change permanent we must add inplace=True otherwise it will not eb reflected in original db

df.drop(1,inplace=True)
print(f"DataFrame after dropping row with index 1 : \n{df}")
```

### Summaries and Group by operations

```python
grouped_mean = df.groupby('department')['salary'].mean()
print(f"Mean salaries of each individual department is : \n{grouped_mean}")
```

```python
# Data types of each column

print(f"Data types of each column are: \n{df.dtypes}")
```

```python
# Describing dataframe

print(f"Statistical summary of dataframe is : \n{df.describe()}")
```
